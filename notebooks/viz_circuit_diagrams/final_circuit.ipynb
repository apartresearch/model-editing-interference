```python
def get_node_name(end_node):
    if isinstance(end_node, int):
        receiver_name = "MLP " + str(end_node)
    elif isinstance(end_node, tuple):
        receiver_name = f"{end_node[0]} , {end_node[1]}"
    else:
        receiver_name = 'resid_post'
    return receiver_name
```

```python
from graphviz import Digraph, Source
from IPython.display import display
from google.colab import files
import pdb

def get_overlap(dot, head_adjList_1, head_adjList_2, head_adjList_3):
    # dot = Digraph()
    # dot.attr(ranksep='0.45', nodesep='0.12')  # Adjust the rankdir, ranksep, and nodesep (defaults 0.75, 0.25)

    if 'resid' in head_adjList_1.keys():
        set1 = set(list(head_adjList_1['resid']))
        set2 = set(list(head_adjList_2['resid']))
        set3 = set(list(head_adjList_3['resid']))
    else:
        set1 = set(list(head_adjList_1.keys()))
        set2 = set(list(head_adjList_2.keys()))
        set3 = set(list(head_adjList_3.keys()))


    # pdb.set_trace()
    if isinstance(list(head_adjList_1.keys())[0], tuple):
        set1 = {(a, b) for a, b, _ in set1}
        set2 = {(a, b) for a, b, _ in set2}
        set3 = {(a, b) for a, b, _ in set3}

    set1 = set( list(set1) + [item for sublist in head_adjList_1.values() for item in sublist] )
    set2 = set( list(set2) + [item for sublist in head_adjList_2.values() for item in sublist] )
    set3 = set( list(set3) + [item for sublist in head_adjList_3.values() for item in sublist] )

    setX = set1.union(set2, set3)
    # print(setX)

    impt_nodes = [(0,1), (1,5), (4,4), (7,11), (9,1)]

    color_dict = {}
    for end_node in setX:
        in_set1 = end_node in set1
        in_set2 = end_node in set2
        in_set3 = end_node in set3

        # node_name = f"{end_node[0]} , {end_node[1]}" # Change to x.y format
        node_name = get_node_name(end_node)

        if in_set1 and not in_set2 and not in_set3:
            dot.node(node_name, color="black", fillcolor="#ffcccb", style='filled',
                     fontsize='20', width='1.1', height='0.6', fontname='Arial bold') #light red
        elif not in_set1 and in_set2 and not in_set3:
            dot.node(node_name, color="black", fillcolor="#ADD8E6", style='filled',
                     fontsize='20', width='1.1', height='0.6', fontname='Arial bold') #light blue
        elif not in_set1 and not in_set2 and in_set3:
            dot.node(node_name, color="black", fillcolor="#FDDC5C", style='filled',
                     fontsize='20', width='1.1', height='0.6', fontname='Arial bold') #light gold
        elif in_set1 and in_set2 and not in_set3:
            dot.node(node_name, color="black", fillcolor="#D8BFD8", style='filled',
                     fontsize='20', width='1.1', height='0.6', fontname='Arial bold') #"#E6E6FA" #light purple
        elif not in_set1 and in_set2 and in_set3:
            dot.node(node_name, color="black", fillcolor="#90EE90", style='filled',
                     fontsize='20', width='1.1', height='0.6', fontname='Arial bold') #light green
        elif in_set1 and not in_set2 and in_set3:
            dot.node(node_name, color="black", fillcolor="#FFD580", style='filled',
                     fontsize='20', width='1.1', height='0.6', fontname='Arial bold')  #light orange
        elif in_set1 and in_set2 and in_set3:
            if end_node in impt_nodes:
                dot.node(node_name, color="black", fillcolor="#EDEDF2", style='filled',
                         penwidth='3', fontsize='20', width='1.1', height='0.6', fontname='Arial bold')
            else:
                dot.node(node_name, fontsize='20', width='1.1', height='0.6', fontname='Arial bold')

    adjList_1_edges = []
    adjList_2_edges = []
    adjList_3_edges = []

    # First, populate the edge lists from all adjacency lists
    for end_node, start_nodes_list in head_adjList_1.items():
        for start_node in start_nodes_list:
            adjList_1_edges.append((start_node, end_node))

    for end_node, start_nodes_list in head_adjList_2.items():
        for start_node in start_nodes_list:
            adjList_2_edges.append((start_node, end_node))

    for end_node, start_nodes_list in head_adjList_3.items():
        for start_node in start_nodes_list:
            adjList_3_edges.append((start_node, end_node))

    # Now, iterate over the union of all edges and draw them according to the rules
    all_edges = set(adjList_1_edges).union(set(adjList_2_edges), set(adjList_3_edges))

    edges_added = []
    for edge in all_edges:
        start_node, end_node = edge
        # start_name = f"{start_node[0]} , {start_node[1]}"
        # end_name = f"{end_node[0]} , {end_node[1]}"
        start_name = get_node_name(start_node)
        end_name = get_node_name(end_node)

        # what = f"9 , 7"
        # if start_name == what or end_name == what:
        #     pdb.set_trace()

        if (start_name, end_name) in edges_added:
            continue

        edges_added.append((start_name, end_name))

        # Edge in all three lists (1, 2, 3)
        if edge in adjList_1_edges and edge in adjList_2_edges and edge in adjList_3_edges:
            dot.edge(str(start_name), str(end_name), color='black', penwidth='2.5') # Bolder gray 707070 # 4.0

        # Edge in lists 1 and 2, but not in 3 (Your new requirement)
        elif edge in adjList_1_edges and edge in adjList_2_edges and edge not in adjList_3_edges:
            dot.edge(str(start_name), str(end_name), color='purple', penwidth='2.5')

        # Edge in lists 1 and 3, but not in 2
        elif edge in adjList_1_edges and edge not in adjList_2_edges and edge in adjList_3_edges:
            dot.edge(str(start_name), str(end_name), color='orange', penwidth='2.5')

        # Edge in lists 2 and 3, but not in 1
        elif edge not in adjList_1_edges and edge in adjList_2_edges and edge in adjList_3_edges:
            dot.edge(str(start_name), str(end_name), color='lightgreen', penwidth='2.5')

        # Edge only in list 1
        elif edge in adjList_1_edges and edge not in adjList_2_edges and edge not in adjList_3_edges:
            dot.edge(str(start_name), str(end_name), color='red', penwidth='2.0')

        # Edge only in list 2
        elif edge not in adjList_1_edges and edge in adjList_2_edges and edge not in adjList_3_edges:
            dot.edge(str(start_name), str(end_name), color='blue', penwidth='2.0')

        # Edge only in list 3
        elif edge not in adjList_1_edges and edge not in adjList_2_edges and edge in adjList_3_edges:
            dot.edge(str(start_name), str(end_name), color='#D4AF37', penwidth='2.0')  # Bolder gold
```

```python
def plot_graph_adjacency(numerals_adjlists, numwords_adjlists, months_adjlists,
                             filename="circuit_graph", highlighted_nodes=None):
    dot = Digraph()
    # dot.attr(ranksep='0.45', nodesep='0.11')  # vert height- ranksep, nodesep- w
    dot.attr(ranksep='0.45', nodesep='0.105')  # vert height- ranksep, nodesep- w

    edges_added = [] # global var so no need pass in
    def loop_adjList(adjList):
        for end_node, start_nodes_list in adjList.items():
            receiver_name = get_node_name(end_node)
            for start in start_nodes_list:
                sender_name = get_node_name(start)
                dot.node(sender_name, color="#ffcccb", style='filled')
                dot.node(receiver_name, color="#ffcccb", style='filled')
                if (sender_name, receiver_name) not in edges_added:
                    dot.edge(sender_name, receiver_name, color = 'red')
                    edges_added.append((sender_name, receiver_name))

    edge_type_list = ['head_to_head_adjList', 'mlp_to_mlp_adjList', 'head_to_mlp_adjList',  'mlp_to_head_adjList', 'heads_to_resid', 'mlps_to_resid']

    for edge_type in edge_type_list:
        adjL_1 = numerals_adjlists[edge_type]
        adjL_2 = numwords_adjlists[edge_type]
        adjL_3 = months_adjlists[edge_type]
        get_overlap(dot, adjL_1, adjL_2, adjL_3)
        # loop_adjList(adjL_1)

    # # Display the graph in Colab
    # display(Source(dot.source))

    dot.format = 'pdf'  # You can change this to 'pdf', 'png', etc. based on your needs
    # dot.format = 'png'  # You can change this to 'pdf', 'svg', etc. based on your needs
    dot.render(filename)
    files.download(filename + ".pdf")
    # files.download(filename + ".png")
```
